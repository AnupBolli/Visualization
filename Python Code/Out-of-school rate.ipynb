```python
import pandas as pd
import numpy as np

df = pd.read_excel('C:\\Users\\welcome\\Downloads\\Out-of-school rate for children of primary school age.xlsx',header = 3)
df1 = pd.read_excel('C:\\Users\\welcome\\Downloads\\Out-of-school rate for adolescents of lower secondary school age.xlsx',header = 3)
df2 = pd.read_excel('C:\\Users\\welcome\\Downloads\\Out-of-school rate for youth of upper secondary school age.xlsx',header = 3)

```

```python
to_drop = ['2012',
           '2013',           
           '2014',
           '2015',
           '2016',            
           '2017',
           '2018']
df.drop(to_drop,inplace=True, axis=1)
df.drop(df.index[0],inplace = True )
df.drop(df.columns[[1,8]], axis = 1, inplace=True)
new_names = {'Time':'Country',
              'Unnamed: 3':'2012',
              'Unnamed: 5':'2013',
              'Unnamed: 7':'2014',
              'Unnamed: 9':'2015',
              'Unnamed: 11':'2016',
              'Unnamed: 13':'2017',
            } 
df.rename(columns = new_names,inplace=True)
df = df.replace('..',0)
for index,row in df.iterrows():
    if row['2017'] != 0:
        df.loc[(index),('Final')] = row['2017']
    elif row['2016'] != 0:
        df.loc[(index),('Final')] = row['2016']   
    elif row['2015'] != 0:
        df.loc[(index),('Final')] = row['2015']  
    elif row['2014'] != 0:
        df.loc[(index),('Final')] = row['2014']          
    elif row['2013'] != 0:
        df.loc[(index),('Final')] = row['2013']  
    else:
        df.loc[(index),('Final')] = row['2012']
df.head(20)
```

```python
df1.drop(df1.index[0],inplace = True )
df1.drop(df1.columns[[1,8]], axis = 1, inplace=True)
df1 = df1.replace('..',0)
df1.rename(columns = {'Time':'Country'},inplace=True)
for index,row in df1.iterrows():
    if row['2017'] != 0:
        df1.loc[(index),('Final')] = row['2017']
    elif row['2016'] != 0:
        df1.loc[(index),('Final')] = row['2016']   
    elif row['2015'] != 0:
        df1.loc[(index),('Final')] = row['2015']  
    elif row['2014'] != 0:
        df1.loc[(index),('Final')] = row['2014']          
    elif row['2013'] != 0:
        df1.loc[(index),('Final')] = row['2013']  
    else:
        df1.loc[(index),('Final')] = row['2012']
     
df1.head(20)
```

```python
df2.drop(df2.index[0],inplace = True )
df2.drop(df2.columns[[1,8]], axis = 1, inplace=True)
df2 = df2.fillna(0)
df2 = df2.replace('..',0)
df2.rename(columns = {'Time':'Country'},inplace=True)
for index,row in df2.iterrows():
    if row['2017'] != 0:
        df2.loc[(index),('Final')] = row['2017']
    elif row['2016'] != 0:
        df2.loc[(index),('Final')] = row['2016']   
    elif row['2015'] != 0:
        df2.loc[(index),('Final')] = row['2015']  
    elif row['2014'] != 0:
        df2.loc[(index),('Final')] = row['2014']          
    elif row['2013'] != 0:
        df2.loc[(index),('Final')] = row['2013']  
    else:
        df2.loc[(index),('Final')] = row['2012']
     
df2.head(20)
```

```python
import plotly.graph_objs as go
import plotly.plotly as py

trace1 = go.Choropleth(
         locationmode = 'country names',
         locations = df['Country'],
         z = df['Final'],
         colorscale = [[0,"rgb(172, 10, 5)"],[0.35,"rgb(190, 60, 40)"],[0.5,"rgb(245, 100, 70)"],\
                     [0.6,"rgb(245, 120, 90)"],[0.7,"rgb(247, 137, 106)"],[1,"rgb(220, 220, 220)"]],
         autocolorscale = False,
         reversescale = True,
         name='PE',
         marker = dict(
            line = dict (
                   color = 'rgb(180,180,180)',
                   width = 0.5
                        ) ),
            colorbar = dict(
                        thickness = 15,
                        len = 0.7,
                        ticksuffix = '%',
                        title = 'Out-of-School Rate',
                        x = 0.9,
                        y = 0.65),
         geo = 'geo1'
        )

trace2 =go.Choropleth(
        locationmode = 'country names',
        locations = df1['Country'],
        z = df1['Final'],
        colorscale = [[0,"rgb(172, 10, 5)"],[0.35,"rgb(190, 60, 40)"],[0.5,"rgb(245, 100, 70)"],\
            [0.6,"rgb(245, 120, 90)"],[0.7,"rgb(247, 137, 106)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        name='LSE',    
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
            colorbar = dict(
                        thickness = 15,
                        len = 0.7,
                        ticksuffix = '%',
                        title = 'Out-of-School Rate',
                        x = 0.9,
                        y = 0.65),
        geo = 'geo1'
         )

trace3 = go.Choropleth(
        locationmode = 'country names',
        locations = df2['Country'],
        z = df2['Final'],
        colorscale = [[0,"rgb(172, 10, 5)"],[0.35,"rgb(190, 60, 40)"],[0.5,"rgb(245, 100, 70)"],
                      [0.6,"rgb(245, 120, 90)"],[0.7,"rgb(247, 137, 106)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        name='HSE',
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
            colorbar = dict(
                        thickness = 15,
                        len = 0.7,
                        ticksuffix = '%',
                        title = 'Out-of-School Rate',
                        x = 0.9,
                        y = 0.65),
        geo = 'geo1'
         )

data = [trace1,trace2,trace3]

updatemenus = list([
    dict(type="buttons",
         active=-1,
         buttons=list([
            dict(label = 'Primary Education',
                 method = 'update',
                 args = [{'visible': [True, False, False]}]),
            dict(label = 'Lower Secondary Education',
                 method = 'update',
                 args = [{'visible': [False,True, False]}]),
            dict(label = 'Higher Secondary Education',
                 method = 'update',
                 args = [{'visible': [False, False,True]}]),
        ]),
         direction = 'up',
         x = 0.16,
         xanchor = 'left',
         y = 0.33,
         yanchor = 'bottom',
         bordercolor = '#41B3A3',
         font = dict(size=11)
    )
])

layout = go.Layout(
    title = 'World''s toughest places to get access to Education <br> \
Source: <a href="https://en.unesco.org/">\
UNESCO</a>',
    geo1 = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'mercator'
        ),
    ),
    legend = dict(
           traceorder = 'reversed'
    ),
    updatemenus=updatemenus    
)

fig = go.Figure(data=data, layout=layout)
py.plot(fig,filename='Out-of-School_Rate.html')
```
