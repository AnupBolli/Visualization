```python
import pandas as pd
import numpy as np

df = pd.read_excel("Scholarships.xls")
df1 = pd.read_excel("Free_Education.xls")

df.head()
```

```python
df.drop(df.index[0],inplace = True )
df = df.fillna(0)
for index,row in df.iterrows():
    if row['2016'] != 0:
        df.loc[(index),('Final')] = row['2016']
    elif row['2015'] != 0:
        df.loc[(index),('Final')] = row['2015']   
    elif row['2014'] != 0:
        df.loc[(index),('Final')] = row['2014']  
    elif row['2013'] != 0:
        df.loc[(index),('Final')] = row['2013']          
    elif row['2012'] != 0:
        df.loc[(index),('Final')] = row['2012']  
    elif row['2011'] != 0:
        df.loc[(index),('Final')] = row['2011']
    else:
        df.loc[(index),('Final')] = row['2010']      

df.head(20)
```

```python
df1.drop(df1.index[0],inplace = True )
df1 = df1.fillna(0)
df1['Final'] = df1['2016']

list = df.Country
df_copy = df1.loc[df1['Country'].isin(list)]
df_copy = df_copy.reset_index()
df_copy.index = df_copy.index + 1
del df_copy['index']
df_copy.head(20)
df1 = df_copy
```

```python
import plotly.graph_objs as go
import plotly.offline as py
from plotly import tools


trace1 = go.Scatter(
    x = df.index,
    y = df.Final,
    mode = 'lines+markers',
    xaxis='x1',
    yaxis='y1',
    name = 'Expenditures spent on Scholarships',
    marker = dict(color = 'rgba(255, 102, 0, 0.7)'),
#                   255, 51, 51, 0.7)'),
#                   255, 102, 0, 0.7)'),
    text = df.Country)

trace2 = go.Scatter(
    x = df1.index,
    y = df1['Final'],
    mode = 'lines+markers',
    xaxis='x2',
    yaxis='y2',
    name = 'Promised No. of years of free education',
    marker = dict(color = 'rgba(102,102,255, 0.6)'),
    text = df1.Country)

layout = go.Layout(
    height=750, 
    width=1220,
    xaxis1=dict(
        title='Countries',
        anchor='y1',        
    ),    
    yaxis1=dict(
        title='Volume of official development assistance flows for scholarships',
        anchor='x1',
    ),
    xaxis2=dict(
        title='Countries',
        domain=[0.5, 1],
        anchor='y2',        
    ),
    yaxis2=dict(
        title='No. of years of free Education',
        domain=[0.5, 1],
        anchor='x2',
    ),
    title = 'UNESCO-Sustainable Development Goal-A pledge to ensure inclusive and equitable quality education for all by 2030'
)



data = [trace1, trace2]

fig = dict(data = data, layout = layout)

py.plot(fig)
```
