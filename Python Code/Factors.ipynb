```python
import plotly.graph_objs as go
import plotly.offline as py
from plotly import tools

trace = go.Bar(
    x=pt.Country,
    y=pt.Final,
    marker=dict(
        color = pt.Final,
        colorscale='Reds')
#         showscale=True)

)

trace1 = go.Bar(
    x=dw.Country,
    y=dw.Final,
    marker=dict(
        color = dw.Final,
        colorscale='Blues',
        reversescale = True)
#         showscale=True)
)

trace2 = go.Bar(
    x=el.Country,
    y=el.Final,
    marker=dict(
        color = el.Final,
        colorscale='Greys',)
#         reversescale = True)
#         showscale=True)
)

trace3 = go.Bar(
    x=ss.Country,
    y=ss.Final,
    marker=dict(
        color = ss.Final,
        colorscale='Greens',
        reversescale = True)
#         showscale=True)
)

fig = tools.make_subplots(rows=2, cols=2, subplot_titles=('Pupil to Teacher ratio', 
                                                          'Access to Drinking Water',
                                                          'Access to Electricity',
                                                          'Access to Single-Sex Basic Sanitation'
                                                          ))
fig.append_trace(trace, 1,1)
fig.append_trace(trace1, 1,2)
fig.append_trace(trace2, 2,1)
fig.append_trace(trace3, 2,2)

fig['layout'].update(height=650, width=1100, title='Factors affecting to Education System', 
                     showlegend=False)                        
py.plot(fig)

# data = [trace,trace1, trace2,trace3]
# layout = go.Layout(barmode = "group")
# fig = go.Figure(data = data, layout = layout)
# py.plot(fig)


```
